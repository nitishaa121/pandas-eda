This is a notebook in which we generate fake data in order to practice simple EDA with pandas and numpy. The data has information on hosts and their event attendance/info, as well as their activity on an sociaal media event posting app. 

```python
import random
import faker
import pandas as pd
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#pip install faker
```

```python

# Create a Faker generator for generating fake data
fake = faker.Faker()
```

```python
# Create an empty DataFrame
data = {
    'User_ID': [],
    'Name': [],
    'Age': [],
    'Gender': [],
    'Location': [],
    'Interests': [],
    'Event_ID': [],
    'Event_Name': [],
    'Event_Type': [],
    'Event_Date': [],
    'Event_Location': [],
    'Event_Attendees': [],
    'Likes': [],
    'Comments': [],
    'Posts': [],
    'Messages_Sent': [],
    'Messages_Received': [],
    'Friends': [],
    'Followers': [],
    'Following': []
}

# Generate data and populate the DataFrame
for _ in range(1000):
    data['User_ID'].append(fake.uuid4())
    data['Name'].append(fake.name())
    data['Age'].append(random.randint(18, 70))
    data['Gender'].append(random.choice(['Male', 'Female', 'Non-Binary']))
    data['Location'].append(fake.city())
    data['Interests'].append(random.choice(['Cooking', 'Sports', 'Gardening', 'Art', 'Music']))
    data['Event_ID'].append(fake.uuid4())
    data['Event_Name'].append(fake.catch_phrase())
    data['Event_Type'].append(random.choice(['Community Gathering', 'Workshop', 'Sports Event', 'Music Concert']))
    data['Event_Date'].append(fake.date_this_year())
    data['Event_Location'].append(fake.city())
    data['Event_Attendees'].append(random.randint(10, 100))
    data['Likes'].append(random.randint(0, 50))
    data['Comments'].append(random.randint(0, 20))
    data['Posts'].append(random.randint(0, 10))
    data['Messages_Sent'].append(random.randint(0, 100))
    data['Messages_Received'].append(random.randint(0, 100))
    data['Friends'].append(random.randint(0, 500))
    data['Followers'].append(random.randint(0, 1000))
    data['Following'].append(random.randint(0, 1000))
```

```python
# Create the DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
df.head()
```

### Most Popular Event Types:

Questions: \
Which event types have the highest average number of attendees? \
Are there any specific types that stand out? \
What is the max number of attendees by event type? Is there a max capacity in place? \
Which top 2 events by type have the most attendees? \
Overall, which event has the third most amount of attendees? (if attendees = 100, 100, 20, 80, 20, 40, the answer is 40 not 80)

```python
# avg number of attendees by event type, none stand out
df.groupby('Event_Type')['Event_Attendees'].mean().sort_values(ascending=False)                                                                                                                          
```

### Event Engagement Ratio Calculation:

Question: \
Calculate the engagement ratio for each event by dividing the total number of likes and comments by the number of attendees and 
add this as a new column. \
Identify the top 5 events with the highest engagement ratio and their event details.

```python
df['Likes_comments_sum'] = df['Likes'] + df['Comments']
```

```python
engagement = round((df.groupby('Event_ID')['Likes_comments_sum'].sum()) / (df.groupby('Event_ID')['Event_Attendees'].sum()),2)
df['Engagement_Ratio'] = df['Event_ID'].map(engagement)
df.head()
```

```python
df = df.drop('Likes_comments_sum', axis=1)
```

```python
df.columns
```

```python
top_5_inds = df['Engagement_Ratio'].nlargest(5).index
df[['Event_ID',
       'Event_Name', 'Event_Type', 'Event_Date', 'Event_Location',
       'Event_Attendees', 'Engagement_Ratio']].iloc[top_5_inds]
```
